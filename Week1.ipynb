## Factors of n

- This is the first program

```python
def factors(n):
  factorList = []
  for i in range(1,n+1):
    if n%i == 0:
      factorList.append(i)
  return factorList
```

```python
factors(7)
```

## Prime

```python
def prime(n):
  return factors(n) == [1, n]
```

Or use this Function for Prime

```python
def prime(n):
  return(len(factors(n))==2)
```

```python
prime(7)
```

```python
primeList = []
for i in range(1, 101):
  if prime(i):
    primeList.append(i)

print(primeList)
```

## Python Recap - GCD - HCF

Greatest Common Divisor (or) Highest Common Factor

gcd(m, n) always exist
- 1 always divides m & n
- always gcd(m, n) <= min(m, n) - Since it has to divide both numbers, it should be minimum or lesser than the numbers
- Prepare a list that divides both m & n from 1 to min(m, n)


```python
def gcd(m, n):
  outputList = []                           # Initialisation
  for i in range(1, min(m, n)+1):           # Control Flow - Loops              # range is always [) so we add +1
    if m%i==0 and n%i==0:                   # Control Flow - Conditionals (if)
      outputList.append(i)                  # append is a method for list type
  return outputList[-1]                     # List runs from 0 to len(n)-1      # In reverse -1 to -len(n)
```

```python
print(gcd(8, 12))
print(gcd(18,9))
```

## Second method for GCD

```python
def gcd(m, n):
  # outputList = []                         # Why do we need to store all the values in array. Last value is suffice
  for i in range(1, min(m, n)+1):           # Control Flow - Loops              # range is always [) so we add +1
    if m%i==0 and n%i==0:                   # Control Flow - Conditionals (if)
      mrcf = i                              # Overwrite the value with the latest value
  return mrcf                               # Can we return something that is not defined? Since gcd always exists as 1. It will return a valid value
```

```python
print(gcd(8, 12))
print(gcd(18,9))
```
